**Outliers**

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile) for each numerical feature
Q1 = df[['Age', 'Income', 'Usage', 'Fitness', 'Miles']].quantile(0.25)
Q3 = df[['Age', 'Income', 'Usage', 'Fitness', 'Miles']].quantile(0.75)

IQR = Q3 - Q1


lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detect outliers: any data points outside these bounds are outliers
outliers = ((df[['Age', 'Income', 'Usage', 'Fitness', 'Miles']] < lower_bound) |
            (df[['Age', 'Income', 'Usage', 'Fitness', 'Miles']] > upper_bound))

# Create a new DataFrame to filter and show rows where any feature is an outlier
outlier_records = df[outliers.any(axis=1)]


outliers_count = outliers.sum()
print("Outliers Count per Feature:")
print(outliers_count)

# Display the records with outliers
print("\nOutlier Records:")
outlier_records
```

**Showing Separately the records of each feature having Outliers**

```python

Q1 = df[['Age', 'Income', 'Usage', 'Fitness', 'Miles']].quantile(0.25)
Q3 = df[['Age', 'Income', 'Usage', 'Fitness', 'Miles']].quantile(0.75)


IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


outlier_details = {}


for col in ['Age', 'Income', 'Usage', 'Fitness', 'Miles']:
    # Identify outliers for the column
    outliers_in_col = df[(df[col] < lower_bound[col]) | (df[col] > upper_bound[col])]

    # Store the outliers and the count for each column
    outlier_details[col] = {
        'outliers': outliers_in_col,
        'count': outliers_in_col.shape[0]
    }

# Display the results for each feature
for col, details in outlier_details.items():
    print(f"\nOutliers in '{col}':")
    print(f"Outlier Count: {details['count']}")
    if details['count'] > 0:
        print(details['outliers'][['Age', 'Income', 'Usage', 'Fitness', 'Miles']])  # Display relevant columns
    else:
        print("No outliers detected.")
```

**Conditional Probabilities**

```python
# Calculate the total number of customers (rows in the dataframe)
total_customers = df.shape[0]

# Count the occurrences of each product in the 'Product' column
product_counts = df['Product'].value_counts()

# Calculate the percentage of customers for each product
product_percentages = (product_counts / total_customers) * 100

product_percentages_with_sign = product_percentages.apply(lambda x: f"{x:.2f}%")

# Display the results
print("Percentage of customers who purchased each product:")
product_percentages_with_sign
```

# Create frequency tables and calculate the percentage as follows
Product – Gender
Percentage of a Male customer purchasing a treadmill
Percentage of a Female customer purchasing KP781 treadmill
Probability of a customer being a Female given that Product is KP281

```python


# 1. Percentage of Male Customers Purchasing a Treadmill
male_customers = df[df['Gender'] == 'Male']
total_customers = df.shape[0]
male_percentage = (male_customers.shape[0] / total_customers) * 100

print(f"Percentage of Male Customers Purchasing a Treadmill: {male_percentage:.2f}%")

# 2. Percentage of Female Customers Purchasing KP781 Treadmill
female_customers_kp781 = df[(df['Gender'] == 'Female') & (df['Product'] == 'KP781')]
total_kp781_customers = df[df['Product'] == 'KP781'].shape[0]
female_percentage_kp781 = (female_customers_kp781.shape[0] / total_kp781_customers) * 100

print(f"Percentage of Female Customers Purchasing KP781 Treadmill: {female_percentage_kp781:.2f}%")

# 3. Conditional Probability: P(Female | Product = KP281)
female_customers_kp281 = df[(df['Gender'] == 'Female') & (df['Product'] == 'KP281')]
total_kp281_customers = df[df['Product'] == 'KP281'].shape[0]
probability_female_given_kp281 = female_customers_kp281.shape[0] / total_kp281_customers

print(f"Probability of a Customer Being Female Given that Product is KP281: {probability_female_given_kp281:.2f}")
```

**Percentage of customers with Age between 20s and 30s**

```python
# Filter customers with age between 20 and 30
age_20s_30s = df[(df['Age'] >= 20) & (df['Age'] < 30)]

# Calculate the total number of customers
total_customers = df.shape[0]

# Calculate the percentage of customers between 20s and 30s
percentage_age_20s_30s = (age_20s_30s.shape[0] / total_customers) * 100


print(f"Percentage of customers with Age between 20s and 30s: {percentage_age_20s_30s:.2f}%")
```

```python

Q1 = df['Income'].quantile(0.25)
Q3 = df['Income'].quantile(0.75)

# low-income and high-income customers defination
low_income_customers = df[df['Income'] < Q1]
high_income_customers = df[df['Income'] > Q3]

# Percentage of Low-Income Customers
low_income_percentage = (low_income_customers.shape[0] / total_customers) * 100
print(f"Percentage of Low-Income Customers Purchasing a Treadmill: {low_income_percentage:.2f}%")

# Percentage of High-Income Customers Purchasing
high_income_customers_kp781 = high_income_customers[high_income_customers['Product'] == 'KP781']
total_high_income_customers = high_income_customers.shape[0]
high_income_percentage_kp781 = (high_income_customers_kp781.shape[0] / total_high_income_customers) * 100
print(f"Percentage of High-Income Customers Purchasing KP781 Treadmill: {high_income_percentage_kp781:.2f}%")

# Conditional Probability: Percentage of High-Income Customers Buying a Treadmill Given that the Product is KP781
high_income_given_kp781 = high_income_customers_kp781.shape[0] / df[df['Product'] == 'KP781'].shape[0]
print(f"Percentage of High-Income Customers Buying Treadmill Given that Product is KP781: {high_income_given_kp781:.2f}")
```
